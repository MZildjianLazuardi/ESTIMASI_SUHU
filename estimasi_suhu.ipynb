# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Total Derajat Suhu Udara Berdasarkan Kelembaban Udara

```python
models = df.groupby('Temperature_C').count()[['Humidity_percent']].sort_values(by='Humidity_percent',ascending=True).reset_index()
models = models.rename(columns={'Humidity_percent':'temperaturOfHumidty'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['Temperature_C'], y=models['temperaturOfHumidty'], color='royalblue')
plt.xticks(rotation=60)
```

# Distribusi Suhu Udara

```python
plt.figure(figsize=(15,5))
sns.distplot(df['Temperature_C'])
```

# seleksi fitur

```python
features = ['UTC', 'Humidity_percent', 'TVOC_ppb', 'eCO2_ppm', 'Raw_H2', 'Raw_Ethanol', 'Pressure_hPa', 'PM1_0', 'PM2_5', 'NC0_5', 'NC1_0', 'NC2_5', 'CNT']
x = df[features]
y = df['Temperature_C']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#UTC', 'Humidity_percent', 'TVOC_ppb', 'eCO2_ppm', 'Raw_H2', 'Raw_Ethanol', 'Pressure_hPa', 'PM1_0', 'PM2_5', 'NC0_5', 'NC1_0', 'NC2_5', 'CNT', 'Fire_Alarm'
input_data = np.array([[1654733454, 52.75, 0, 400, 12563, 19658, 939.798, 0.01, 0.02, 0.01, 0.025, 0.015, 123]])

prediction = lr.predict(input_data)
print('Estimasi Suhu Udara(celcius) :', prediction, 'Celcius')
```
